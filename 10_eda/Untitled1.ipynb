```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
nucinfo=pd.read_csv("nuclei_features.csv")
```

```python
nucinfo["celltype"] = nucinfo["image_key"].str.split("_").str[0]
```

```python
stats = (
    nucinfo.groupby("celltype")["equivalent_radius_px"]
           .agg(["count", "mean", "median", "std", "min", "max"])
           .sort_values("median", ascending=False)
)
print(stats.head())

import matplotlib.pyplot as plt

nucinfo.boxplot(column="equivalent_radius_px", by="celltype", rot=90, figsize=(10,5))
plt.ylabel("equivalent_radius_px")
plt.title("Nuclear Radius by celltype")
plt.suptitle("")
plt.tight_layout()
plt.show()
```

```python
cellinfo=pd.read_csv("cell_features.csv")
cellinfo["celltype"] = cellinfo["image_key"].str.split("_").str[0]
```

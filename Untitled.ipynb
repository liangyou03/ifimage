```python
import logging
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from utils import Sample, SampleDataset, ensure_dir

# --- basic logger setup (prints to stdout) ---
logger = logging.getLogger("plot_scribbles")
if not logger.handlers:
    handler = logging.StreamHandler()
    fmt = logging.Formatter("[%(levelname)s] %(message)s")
    handler.setFormatter(fmt)
    logger.addHandler(handler)
logger.setLevel(logging.INFO)


def _save(fig, out_path: Path):
    """Save figure with tight bbox and minimal padding."""
    fig.tight_layout()
    fig.savefig(out_path, dpi=fig.dpi, bbox_inches="tight", pad_inches=0.02)
    plt.close(fig)


def plot_sample_comparisons(sample: Sample, out_dir: Path, dpi: int = 300, logger: logging.Logger = logger):
    """
    For one Sample, conditionally generate figures ONLY if the corresponding manual scribble exists.
    - Figure A: DAPI vs nucleus scribble (requires nuc_scribble)
    - Figure B: (Marker + DAPI) vs cell scribble (requires cell_chan AND marker_scribble)

    Returns a dict of {"figure_name": saved_path} for figures actually written.
    """
    sample.load_images()
    sample.load_scribbles()
    ensure_dir(out_dir)
    saved = {}

    # ---------- A) DAPI vs nucleus scribble ----------
    if sample.nuc_scribble is None:
        logger.info(f"[{sample.base}] skip DAPI-vs-nuc: no nuc_scribble")
    else:
        if sample.nuc_chan is None:
            logger.info(f"[{sample.base}] skip DAPI-vs-nuc: no nuc_chan (DAPI image missing)")
        elif sample.nuc_scribble.shape != sample.nuc_chan.shape:
            logger.warning(f"[{sample.base}] skip DAPI-vs-nuc: shape mismatch scribble {sample.nuc_scribble.shape} vs dapi {sample.nuc_chan.shape}")
        else:
            fig, axs = plt.subplots(1, 2, figsize=(12, 6), dpi=dpi)
            axs[0].imshow(sample.nuc_chan, cmap="gray")
            axs[0].set_title("DAPI channel")
            axs[0].axis("off")

            axs[1].imshow(sample.nuc_chan, cmap="gray")
            axs[1].imshow(sample.nuc_scribble, cmap="jet", alpha=0.5)
            axs[1].set_title("Nucleus scribble overlay")
            axs[1].axis("off")

            out_path = out_dir / f"{sample.base}_dapi_vs_nuc.png"
            _save(fig, out_path)
            saved["dapi_vs_nuc"] = str(out_path)
            logger.info(f"[{sample.base}] saved {out_path.name}")

    # ---------- B) (Marker + DAPI) vs cell scribble ----------
    need = []
    if sample.cell_chan is None:
        need.append("marker channel")
    if sample.marker_scribble is None:
        need.append("cell scribble")
    if need:
        logger.info(f"[{sample.base}] skip marker+dapi-vs-cell: missing {', '.join(need)}")
    else:
        # Check shapes
        if sample.cell_chan.shape != sample.nuc_chan.shape:
            logger.warning(f"[{sample.base}] skip marker+dapi-vs-cell: channel shape mismatch marker {sample.cell_chan.shape} vs dapi {sample.nuc_chan.shape}")
        elif sample.marker_scribble.shape != sample.cell_chan.shape:
            logger.warning(f"[{sample.base}] skip marker+dapi-vs-cell: scribble {sample.marker_scribble.shape} vs channels {sample.cell_chan.shape}")
        else:
            # Compose RGB: Marker -> R, DAPI -> B
            rgb = np.zeros((*sample.cell_chan.shape, 3), dtype=np.float32)
            rgb[..., 0] = sample.cell_chan  # red = marker
            rgb[..., 2] = sample.nuc_chan   # blue = dapi

            fig, axs = plt.subplots(1, 2, figsize=(12, 6), dpi=dpi)
            axs[0].imshow(rgb)
            axs[0].set_title("Marker + DAPI composite")
            axs[0].axis("off")

            axs[1].imshow(rgb)
            axs[1].imshow(sample.marker_scribble, cmap="jet", alpha=0.5)
            axs[1].set_title("Cell scribble overlay")
            axs[1].axis("off")

            out_path = out_dir / f"{sample.base}_marker_dapi_vs_cell.png"
            _save(fig, out_path)
            saved["marker_dapi_vs_cell"] = str(out_path)
            logger.info(f"[{sample.base}] saved {out_path.name}")

    return saved


def batch_plot(data_dir: Path, out_dir: Path, dpi: int = 300):
    """
    Iterate all samples and generate figures when eligible.
    Prints a compact summary at the end.
    """
    ds = SampleDataset(Path(data_dir))
    ensure_dir(out_dir)

    total_samples = 0
    total_figs = 0
    for sample in ds:
        total_samples += 1
        saved = plot_sample_comparisons(sample, out_dir, dpi=dpi, logger=logger)
        total_figs += len(saved)

    logger.info(f"Done. {total_figs} figures written from {total_samples} samples.")


# Example usage:
# batch_plot(Path("your_data_dir"), Path("plots"), dpi=400)
```
